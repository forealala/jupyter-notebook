```python
a2 = a1.dropna()
a2
```

```python
x1 = np.array(a2['x'])
y1 = np.array(a2['y'])
```

```python
x1.shape
```

```python
x2 = x1.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x2,y1)
```

```python
a = lr.coef_
b = lr.intercept_
a, b
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))
plt.plot(x1,y1,'o')
plt.plot(x2, lr.predict(x2))
x = np.linspace(0,100,10000)
plt.plot(x, a*x+b, 'r')
```
