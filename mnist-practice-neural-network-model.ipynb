```python
from tensorflow.keras.datasets.mnist import load_data
(train_x, train_y), (test_x, test_y) = load_data()
```

```python
train_x.shape, train_y.shape
test_x.shape, test_y.shape
```

```python
from PIL import Image
img = train_x[0]

import matplotlib.pyplot as plt
img1 = Image.fromarray(img, mode='L')
plt.imshow(img1)
train_y[0]
```

```python
train_x1 = train_x.reshape(60000,-1)
test_x1 = test_x.reshape(10000, -1)
```

```python
train_x2 = train_x1/255
test_x2 = test_x1/255
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
md = Sequential()
md.add(Dense(128, activation='relu', input_shape = (28*28,)))
md.add(Dense(64, activation='relu'))
md.add(Dense(10, activation='softmax'))
md.summary()
```

```python
md.compile(loss='sparse_categorical_crossentropy',
          optimizer='sgd',metrics='acc')
```

```python
hist = md.fit(train_x2, train_y, epochs=50, batch_size=64, validation_split=0.2)
```

```python
acc = hist.history['acc']
val_acc = hist.history['val_acc']
epoch = np.arange(1,len(acc)+1)

plt.figure(figsize = (10,8))
plt.xlim(0, len(acc)+1)
plt.ylim(0.85,1.0)
plt.plot(epoch,acc,'b',label='acc')
plt.plot(epoch,val_acc,'g',label='val_acc')
plt.legend()
```

```python
md.evaluate(test_x2,test_y)
```

```python
# dropout 적용 => 모델 재설정
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

md = Sequential()
md.add(Dense(128, activation='relu', input_shape = (28*28,)))
md.add(Dense(64, activation='relu'))
md.add(Dropout(0.5))
md.add(Dense(10, activation='softmax'))
md.summary()
```

```python
md.compile(loss='sparse_categorical_crossentropy',
          optimizer='sgd',metrics='acc')

hist = md.fit(train_x2, train_y, epochs=50, batch_size=64, validation_split=0.2)
```
